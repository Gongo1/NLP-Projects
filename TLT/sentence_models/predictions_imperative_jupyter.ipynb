## Get the predictions to test the accuracy of the model
After we have trained out model we can test it to see how it does in a real scenarion

```python
import xgboost as xgb
import pandas as pd
```

```python
model = xgb.Booster(model_file = "imperative.model") #Provide: name of the .model file
data = pd.read_csv('data/validation_data.csv') #Provide: name of the validation dataset file
                                                #Important to note that the validation dataset has to be formatted the same way as the training dataset

y_col = 'label' #Provide: name of the label column
x_col = 'text' #Provide: name of the data column
text = data[[x_col]]
y_val = data[[y_col]]
x_val = data.drop(y_col, axis=1)
x_val = data.drop(x_col, axis=1)
```

```python
dtest = xgb.DMatrix(x_val) #validation dataset
                            #A DMatrix is a data format used by XGBoost
    
print('Initializing Inference... \n')
preds = model.predict(dtest)
print('Inference Finished.')
    
preds = pd.DataFrame(preds)
preds.columns = ['Prediction']
```
